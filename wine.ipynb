```python
quality_counts = wi["quality"].value_counts()
quality_counts
```

```python
plt.figure(figsize = (8,6))
plt.bar(quality_counts.index,quality_counts,color = "pink")
plt.title("count plot of Quality")
plt.xlabel("Quality")
plt.ylable("Count")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x ="quality" , y="alcohol" ,data=wi,palette='viridis')
plt.title("Swarm plot for Quality and alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
wi["quality"] = wi["quality"].astype(str)
plt.figure(figsize=(10,8))
sns.violinplot(x ="quality" , y="alcohol" ,data=wi,palette='viridis')
plt.title("Violin plot for Quality and alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
sns.boxplot(x ="quality" , y="alcohol" ,data=wi)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(wi.corr(),annot=True,fmt=".2f",cmap="Pastel2",linewidths=2)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x ="quality" , y="citric acid" ,data=wi)
plt.show()
```

```python
plt.subplots(figsize = (10,4))
plt.bar(wi.quality,wi.chlorides,width=0.6,color="red")
plt.title("No of the Deaths Per Years ")
plt.xlabel("Quality",fontsize = 14)
plt.ylabel("chlorides",fontsize = 14)
plt.show()
```
